**1. Criação dos Métodos**

```python
df.head()
```

```python
def empty_values(dataframe):
  """Retorna soma dos valores nulos"""
  return dataframe.isnull().sum()

def remove_empty_values(dataframe, name_column):
  """Remove valores vazios"""
  return dataframe[dataframe[name_column].notna()]

def dataframe_grouped(dataframe, column1, column2, column_agreg, type_, new_column):
  """Realiza agregação de um dataframe"""

  dataframe_tranformed = dataframe.groupby([column1, column2]).agg({column_agreg: [type_]})
  dataframe_tranformed.columns = [new_column]

  return dataframe_tranformed.reset_index()
```

```python
empty_values(df)
```

```python
df_filter = remove_empty_values(df, 'last_review')
empty_values(df_filter)
```

```python
grouped_multiple = dataframe_grouped(df_filter,   
                                     'room_type', 
                                     'last_review', 'price', 'mean', 'price_mean')
```

```python
grouped_multiple['last_review'] = pd.to_datetime(grouped_multiple['last_review'], format='%Y-%m-%d')
grouped_multiple.head()
```

```python
grouped_multiple['quarter'] = grouped_multiple['last_review'].dt.quarter
grouped_multiple.head()
```

```python
grouped_multiple.groupby(['room_type']).size()
```

```python
# 1- Gráficos de linhas de variáveis numéricas (médias por categorias no tempo) (PD)
```

**2. Análise dos gráficos**

```python


font1 = {'family':'serif','color':'black','size':15}
font2 = {'family':'serif','color':'darkred','size':15}


ax1 = grouped_multiple[grouped_multiple['room_type'] == 'Entire home/apt'].plot(x='last_review', y='price_mean', c='green', label = 'Entire home/apt',)
ax2 = grouped_multiple[grouped_multiple['room_type'] == 'Private room'].plot(x='last_review', y='price_mean', c='red', ax=ax1, label = 'Private room')
ax3 = grouped_multiple[grouped_multiple['room_type'] == 'Shared room'].plot(x='last_review', y='price_mean', c='blue', ax=ax1, label='Shared room')


plt.title("Evolução segundo tipo de quartos Airbnb", loc = 'left', fontdict = font1)
plt.xlabel("Ano")
plt.ylabel("Preço Médio")

plt.show()
```

```python
sub_grouped_multiple=grouped_multiple[grouped_multiple.price_mean < 1000]

font1 = {'family':'serif','color':'black','size':15}
font2 = {'family':'serif','color':'darkred','size':15}


ax1 = sub_grouped_multiple[sub_grouped_multiple['room_type'] == 'Entire home/apt'].plot(x='last_review', y='price_mean', c='green', label = 'Entire home/apt',)
ax2 = sub_grouped_multiple[sub_grouped_multiple['room_type'] == 'Private room'].plot(x='last_review', y='price_mean', c='red', ax=ax1, label = 'Private room')
ax3 = sub_grouped_multiple[sub_grouped_multiple['room_type'] == 'Shared room'].plot(x='last_review', y='price_mean', c='blue', ax=ax1, label='Shared room')


plt.title("Evolução segundo tipo de quartos Airbnb - < 1000", loc = 'left', fontdict = font1)
plt.xlabel("Ano", fontdict = font2)
plt.ylabel("Preço Médio", fontdict = font2)

plt.show()
```

```python
# 2- Gráficos de setores (pizza) com algumas variáveis categóricas (diferentes das já analisadas) (PX)

fig = px.pie(df, values='number_of_reviews', 
             names='neighbourhood_group', 
             width=600, 
             height=500, 
             color_discrete_sequence=px.colors.sequential.RdBu,
             title="Participação relativa segundo grupos de bairros")

fig.show()
```

```python
# 3- Gráficos de barras horizontais com alguma variável dependente versus média da variável resposta (PLT)
```

```python
plt.figure(figsize=(10,5))
df_type = df.pivot_table(index='neighbourhood_group', values='minimum_nights', aggfunc='mean').sort_values('minimum_nights', ascending=False)
plt.bar(df_type.index, df_type['minimum_nights'], color=['black', 'red', 'green', 'blue', 'cyan']);
plt.xlabel("Grupo de Bairros", size=15)
plt.ylabel("Número de Noites Mínimas", size=15)
plt.title("Grupos de Bairro segundo média de noites mínimas", size=18)
```

```python
# 4- Pairplot e heatmap para análise visual da correlação das variáveis independentes (SNS)


plt.figure(figsize=(10, 5))

# Calculando correlação
corr = df.corr()

# Plotando o método
sns.heatmap(corr, 
        xticklabels=corr.columns,
        yticklabels=corr.columns, cmap="RdBu_r", annot=True)

plt.title('Correlação Variáveis Airbnb', fontsize = 20) # title with fontsize 20
plt.xlabel('Variáveis', fontsize = 15) # x-axis label with fontsize 15
plt.ylabel('Variáveis', fontsize = 15) # y-axis label with fontsize 15

plt.show()
```

```python
plt.figure(figsize=(10, 5))

sub_df=df[df.price < 400]

sns.boxplot( y=sub_df["price"], x=sub_df["neighbourhood_group"] );
plt.title('Boxplot classes de bairros - preço < 400', fontsize = 20)
plt.xlabel('Bairros', fontsize = 15) 
plt.ylabel('Preços', fontsize = 15)

plt.show()
```
